```
import pandas as pd
import numpy as np


df=pd.read_csv('aaaa.csv',header=None)
df
```

```
x=df.shape[0]
y=df.shape[1]
```

```
df=df.drop(df.index[0])
del df[0]
df
```

```
X=df.shape[0]
Y=df.shape[1]
```

```
df.iloc[:X,2:3]=df.iloc[:X,2:3].replace("Asia","4")
df.iloc[:X,2:3]=df.iloc[:X,2:3].replace("Africa","3")
df.iloc[:X,2:3]=df.iloc[:X,2:3].replace("Europe","2")
df.iloc[:X,2:3]=df.iloc[:X,2:3].replace("Americas","1")
df.iloc[:X,3:4]=df.iloc[:X,3:4].replace("yes","1")
df.iloc[:X,3:4]=df.iloc[:X,3:4].replace("no","0")
df
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[0:X,0:3], df.iloc[0:X,3:4], test_size=0.30)
```

```
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

ID3

```
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion="entropy")
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

LOGISTIC REGRESSION

```
from sklearn.linear_model import LogisticRegression
classifier =  LogisticRegression(tol=0.001)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

RANDOM FOREST

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

NUERAL NETWORK

```
from sklearn.neural_network import MLPClassifier
classifier = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
